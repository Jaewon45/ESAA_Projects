# 라벨 분리

- 개별 운동을 라밸로 삼아 데이터분석을 진행할 수 있도록, 문자열 형태로 되어있는 운동처방 칼럼 ('MVM_PRSCRPTN_CN')을 분리해야하며, 해당 개별 운동들을 칼럼명으로 삼은 각각의 데이터프레임을 만들어야 한다.
- 이를 위해 문자열을 크게 준비운동, 본운동, 마무리운동으로 분리하고 각 단계를 구성하는 운동들을 각각 분리시켜 리스트로 만든 후, 리스트의 각 요소들을 칼럼명으로 삼고 해당 운동을 처방한 경우는 1, 아닌 경우는 0을 데이터값으로 가지고 있는 열을 추가하도록 코드를 작성했다.

```python
data_mvm = data['MVM_PRSCRPTN_CN'].dropna(axis=0)
```

```python
data_ispre = data_mvm[data_mvm.str.startswith('준')]  #준비운동 포함
data_nopre = data_mvm[data_mvm.str.startswith('준')==False] # 준비운동 포함 안 함
```

```python
ispre_nomain = data_ispre.str.split('본운동:',expand=True).reset_index(drop=True)[:][0] #마무리운동 포함하고 있을 수도있는 데이터
ispre_ismain = data_ispre.str.split('본운동:',expand=True).reset_index(drop=True)[:][1]#본운동 o, 마무리운동 포함하고 있을 수도있는 데이터
```

```python
ispre_ismain.isna().sum() #준비 운동만 있는 데이터
```

```python
ispre_ismain.dropna(axis=0, inplace=True) #준비 운동만 있는 데이터 제거
```

```python
ispre_ismain.isna().sum()
```

# 데이터 분석
- 데이터분석은 크게 군집화, 분석 두가지로 진행했다.
 - 군집화의 경우 운동처방데이터를 가장 적절하게 여러 군집으로 클러스터링한 군집화 기법을 선택, 새로운 운동측정변수를 입력했을 때 입력한 정보와 가장 근접한 군집에서 처방받은 운동들을 원하는 크기(비율)만큼 출력하는 것을 목표로 한다.
 - 분류의 경우 데이터를 훈련셋과 테스트셋으로 나누어, 각각의 개별 운동의 처방여부 (0, 1)에 대한 여러가지 모델을 활용한 분류 및 하이퍼파라미터 튜닝을 진행해 가장 성능이 좋은 분류기를 선택한다. 이후 새로운 운동측정변수를 입력했을 때 적합시킨 모델들로 처방받게 될 운동들을 예측하여 처방받게 될 (라밸값이 1인) 운동들을 모아 출력하는 것을 목표로 한다.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import math

from itertools import combinations
```

```python
def visualize_silhouette(cluster_lists, X_features):     
    # 입력값으로 클러스터링 갯수들을 리스트로 받아서, 각 갯수별로 클러스터링을 적용하고 실루엣 개수를 구함
    n_cols = len(cluster_lists)
   
    # plt.subplots()으로 리스트에 기재된 클러스터링 수만큼의 sub figures를 가지는 axs 생성 
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)
   
    # 리스트에 기재된 클러스터링 갯수들을 차례로 iteration 수행하면서 실루엣 개수 시각화
    for ind, n_cluster in enumerate(cluster_lists):
       
        # KMeans 클러스터링 수행하고, 실루엣 스코어와 개별 데이터의 실루엣 값 계산. 
        clusterer = KMeans(n_clusters = n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(X_features)
       
        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)
       
        y_lower = 10
        axs[ind].set_title('Number of Cluster : '+ str(n_cluster)+'\n' \
                          'Silhouette Score :' + str(round(sil_avg,3)) )
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])  # Clear the yaxis labels / ticks
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
       
        # 클러스터링 갯수별로 fill_betweenx( )형태의 막대 그래프 표현. 
        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels==i]
            ith_cluster_sil_values.sort()
           
            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i
           
            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values, \
                                facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10

        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")
```

```python
def visualize_silhouette_eps_6(minsamples_lists, X_features): 

    from sklearn.datasets import make_blobs
    from sklearn.cluster import DBSCAN
    from sklearn.metrics import silhouette_samples, silhouette_score

    import matplotlib.pyplot as plt
    import matplotlib.cm as cm
    import math

    # 입력값으로 min_samples 개수들을 리스트로 받아서, 각 min_samples별로 클러스터링을 적용하고 실루엣 개수를 구함
    n_cols = len(minsamples_lists)

    # plt.subplots()으로 리스트에 기재된 클러스터링 수만큼의 sub figures를 가지는 axs 생성 
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)

    # 리스트에 기재된 클러스터링 갯수들을 차례로 iteration 수행하면서 실루엣 개수 시각화
    for ind, minsample in enumerate(minsamples_lists):

        # DBSCAN 클러스터링 수행하고, 실루엣 스코어와 개별 데이터의 실루엣 값 계산. 
        clusterer = DBSCAN(eps = 0.6, min_samples = minsample, metric = "euclidean")
        cluster_labels = clusterer.fit_predict(X_features)
        n_cluster = len(pd.Series(cluster_labels).unique())

        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)

        y_lower = 10
        axs[ind].set_title('Number of Cluster : '+ str(n_cluster)+'\n' \
                          'Silhouette Score :' + str(round(sil_avg,3)) )
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])  # Clear the yaxis labels / ticks
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])

        # 클러스터링 갯수별로 fill_betweenx( )형태의 막대 그래프 표현. 
        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels==i]
            ith_cluster_sil_values.sort()

            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values, \
                                facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10

        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")
```

 최종적으로, 사용자의 체력 측정 값을 리스트 형태로 입력하면 해당 데이터와 군집의 중심이 가장 가까운 군집의 번호를 리턴해주는 함수이다. (find_closest_cluster())

```python
def find_closest_cluster(user_input):
  user_input=pd.DataFrame({'MESURE_AGE_CO':[0],
                         'SEXDSTN_FLAG_CD':[0],
                         'MESURE_IEM_001_VALUE':[0],
                         'MESURE_IEM_002_VALUE':[0],
                         'MESURE_IEM_003_VALUE':[0],
                         'MESURE_IEM_005_VALUE':[0],
                         'MESURE_IEM_006_VALUE':[0],
                         'MESURE_IEM_012_VALUE':[0],
                         'MESURE_IEM_018_VALUE':[0]})

  #리스트 형태로 입력받은 User data를 데이터 프레임 형태의 user_input에 저장
  for i in range(len(user_input_list)):
    user_input.iloc[0,i]=user_input_list[i]
  
  # 성별 제외 변수 standardiztion
  user_data_stand=standardization(user_input)

  # 성별 변수 더미화 
  user_data_stand=user_input_likeGetDummies(user_data_stand)

  # 더미화 된 성별 변수 표준화
  user_data_stand['SEXDSTN_FLAG_CD_F']=(user_data_stand['SEXDSTN_FLAG_CD_F']-np.mean(DF_2022['SEXDSTN_FLAG_CD_F']))/np.std(DF_2022['SEXDSTN_FLAG_CD_F'])
  user_data_stand['SEXDSTN_FLAG_CD_M']=(user_data_stand['SEXDSTN_FLAG_CD_M']-np.mean(DF_2022['SEXDSTN_FLAG_CD_M']))/np.std(DF_2022['SEXDSTN_FLAG_CD_M'])

  # 표준화된 user data와 가장 가까운 군집 출력
  closest_cluster=min_dist_index(user_data_stand,kmeans.cluster_centers_)
  return closest_cluster
```

```python
col = ['MESURE_AGE_CO','SEXDSTN_FLAG_CD','MESURE_IEM_001_VALUE','MESURE_IEM_002_VALUE','MESURE_IEM_003_VALUE','MESURE_IEM_005_VALUE','MESURE_IEM_006_VALUE','MESURE_IEM_012_VALUE','MESURE_IEM_018_VALUE']
data_pre_selected = data_pre[col]
data_main_selected = data_main[col]
data_end_selected = data_end[col]

# 준비운동 분류용 데이터프레임 생성
for i in range (1,11):  
  globals()['data_pre'+str(i)] = pd.concat([data_pre_selected, pd.DataFrame(data_pre.iloc[:,(i-11)])], axis=1)

# 본운동 분류용 데이터프레임 생성
for i in range (1,37):  
  globals()['data_main'+str(i)] = pd.concat([data_main_selected, pd.DataFrame(data_main.iloc[:,(i-37)])], axis=1)

# 마무리운동 분류용 데이터프레임 생성
for i in range (1,9):  
  globals()['data_end'+str(i)] = pd.concat([data_end_selected, pd.DataFrame(data_end.iloc[:,(i-9)])], axis=1)
```

```python
# 분류를 위한 데이터전처리용 함수 생성
def preprocessing(df):
  import pandas as pd
  from sklearn.model_selection import train_test_split
  from imblearn.over_sampling import SMOTE

  # 1) 원핫 인코딩 
  df_rev = pd.get_dummies(data=df)

  # 2) X, y 데이터로 분리 
  temp = list(range(11)); del temp[-3]
  X = df_rev.iloc[:,temp]
  y = df_rev.iloc[:,-3]

  # 3) 학습/테스트용 데이터세트 분리
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

  # 4) SMOTE를 이용한 오버샘플링
  smote = SMOTE(random_state=42)
  X_train_over, y_train_over = smote.fit_resample(X_train, y_train)

  return (X_train, X_test, y_train, y_test, X_train_over, y_train_over)
```

```python
## 오차행렬 / 정확도 / 정밀도 / 재현율 출력 함수
def get_clf_eval(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    print(model,"==================================")
    print('오차행렬')
    print(confusion)
    print('정확도 : {0:.4f}, 정밀도 : {1:.4f}, 재현율:{2:.4f}'.format(accuracy, precision, recall))
    return (accuracy, precision, recall, pred)
```

```python
## max_depth별 랜덤포레스트 성능평가 출력 함수
def get_rf_eval(X_train, y_train, X_test, y_test):
  from sklearn.tree import DecisionTreeClassifier
  from sklearn.ensemble import RandomForestClassifier

  rf_md5 = RandomForestClassifier(max_depth=5, n_estimators=400, random_state=42)
  get_clf_eval(X_train, y_train, X_test, y_test, rf_md5)

  rf_md8 = RandomForestClassifier(max_depth=8, n_estimators=400, random_state=42)
  get_clf_eval(X_train, y_train, X_test, y_test, rf_md8)

  rf_md11 = RandomForestClassifier(max_depth=11, n_estimators=400, random_state=42)
  get_clf_eval(X_train, y_train, X_test, y_test, rf_md11)

  rf_md14 = RandomForestClassifier(max_depth=14, n_estimators=400, random_state=42)
  get_clf_eval(X_train, y_train, X_test, y_test, rf_md14)

  rf_md17 = RandomForestClassifier(max_depth=17, n_estimators=400, random_state=42)
  get_clf_eval(X_train, y_train, X_test, y_test, rf_md17)
```

- 최적화된 하이퍼파라미터에 대한 성능평가를 위해 정확도(accuracy), 재현율(recall), 정밀도(precision) 및 오분류표를 출력한 결과, Random Forest를 제외한 XGBoost, lightGBM의 경우는 정확도가 높지만 재현율이 상대적으로 많이 저조해 true y가 1인 경우에 대한 올바른 분류가 적은 것을 확인할 수 있다.따라서 재현율이 비교적 높은 랜덤포레스트를 선택했다.
- 랜덤포레스트에서 하이퍼파라미터를 튜닝해 성능평가를 진행한 결과, max_depth가 높을수록 정확도가 증가하지만 그만큼 재현율은 감소하며, max_depth가 지나치게 크면 과적합의 위험 또한 증가하기 때문에 정확도가 일정 값 이상 (>0.6)인 가장 작은 값을 max_depth로 설정하도록하는 함수를 생성해 해당 max_depth와 n_estimators=400을 하이퍼파라미터로 갖는 랜덤 포레스트로 분류를 진행했다.

```python
## 정확도>0.6인 최저 max_depth의 predict값을 새로운 칼럼으로 할당 / 데이터프레임 구성 및 성능평가 출력 함수
def rf_pred_eval(X_train, y_train, X_test, y_test):
  from sklearn.tree import DecisionTreeClassifier
  from sklearn.ensemble import RandomForestClassifier
  from sklearn.metrics import accuracy_score, precision_score, recall_score

  for i in range (3, 10):
    rf_md = RandomForestClassifier(max_depth=i, n_estimators=400, random_state=42)
    rf_md.fit(X_train, y_train)
    pred = rf_md.predict(X_test)
    accuracy = accuracy_score(y_test, pred)
    if accuracy > 0.6:
      X_test[y_test.name]=pred
      org[y_test.name]=pred
      precision = precision_score(y_test, pred)
      recall = recall_score(y_test, pred)
      clf_accuracy.append(accuracy)
      clf_recall.append(recall)
      clf_precision.append(precision)
      print(rf_md)
      print('정확도 : {0:.4f}, 정밀도 : {1:.4f}, 재현율:{2:.4f}'.format(accuracy, precision, recall))
      break
```

```python
#운동리스트 생성 함수
def mvm_list(data):
  data_mvm = data['MVM_PRSCRPTN_CN'].dropna(axis=0)

  data_ispre = data_mvm[data_mvm.str.startswith('준')]  #준비운동 포함
  data_nopre = data_mvm[data_mvm.str.startswith('준')==False] # 준비운동 포함 안 함

  ispre_nomain = data_ispre.str.split('본운동:',expand=True).reset_index(drop=True)[:][0]
  ispre_ismain = data_ispre.str.split('본운동:',expand=True).reset_index(drop=True)[:][1]
  ispre_ismain.dropna(axis=0, inplace=True)

  ispre_nomain_noend = ispre_nomain.str.split('마무리운동:',expand=True).reset_index(drop=True)[:][0] 
  ispre_nomain_isend = ispre_nomain.str.split('마무리운동:',expand=True).reset_index(drop=True)[:][1]
  ispre_nomain_isend.dropna(axis=0, inplace=True)

  ispre_ismain_noend = ispre_ismain.str.split('마무리운동:',expand=True).reset_index(drop=True)[:][0] 
  ispre_ismain_isend = ispre_ismain.str.split('마무리운동:',expand=True).reset_index(drop=True)[:][1] 
  ispre_ismain_isend.dropna(axis=0, inplace=True)

  nopre_noend = data_nopre.str.split('마무리운동:',expand=True).reset_index(drop=True)[:][0]
  nopre_isend = data_nopre.str.split('마무리운동:',expand=True).reset_index(drop=True)[:][1]
  nopre_isend.dropna(axis=0,inplace=True)

  #준비운동 리스트 만들기
  ispre_nomain_noend = ispre_nomain_noend.str.replace('준비운동:','')
  ispre_nomain_noend = ispre_nomain_noend.str.replace(' / ','')
  pre_data = ispre_nomain_noend.str.split(',')
  pre_list = []
  for i in range(pre_data.shape[0]):
    for j in range(len(pre_data[i])):
      pre_list.append(pre_data[i][j])
  pre_list = pd.DataFrame(pre_list)

  #본운동 리스트 만들기
  nopre_noend = nopre_noend.str.replace('본운동:','')
  nopre_noend = nopre_noend.str.replace(' / ','')
  nopre_noend = nopre_noend.str.split(',')
  ispre_ismain_noend = ispre_ismain_noend.str.replace(' / ','')
  ispre_ismain_noend = ispre_ismain_noend.str.split(',')
  main_data = pd.concat([nopre_noend, ispre_ismain_noend],axis=0)
  main_data.reset_index(drop=True, inplace=True)
  main_list = []
  for i in range(main_data.shape[0]):
    for j in range(len(main_data[i])):
        main_list.append(main_data[i][j])
  main_list = pd.DataFrame(main_list)

  #마무리운동 리스트 만들기
  ispre_nomain_isend = ispre_nomain_isend.str.split(',')
  ispre_ismain_isend = ispre_ismain_isend.str.split(',')
  nopre_isend = nopre_isend.str.split(',')
  end_data = pd.concat([ispre_nomain_isend, ispre_ismain_isend, nopre_isend], axis=0)
  end_data.reset_index(drop=True, inplace=True)
  end_list = []
  for i in range(end_data.shape[0]):
    for j in range(len(end_data[i])):
        end_list.append(end_data[i][j])
  end_list = pd.DataFrame(end_list)
  
  return pre_list, main_list, end_list
```

```python
#클러스터별 운동 리스트 생성
c0_pre, c0_main, c0_end = mvm_list(df_2022_c0)
c1_pre, c1_main, c1_end = mvm_list(df_2022_c1)
c2_pre, c2_main, c2_end = mvm_list(df_2022_c2)
c3_pre, c3_main, c3_end = mvm_list(df_2022_c3)
c4_pre, c4_main, c4_end = mvm_list(df_2022_c4)
```

## 분류 결과 운동 리스트 출력

#### a. 준비운동 분류 함수

```python
## 준비운동 분류 함수 (데이터프레임/리스트 형태로 결과 반환)
def pre_df_list(input_list):
  
  # 1)리스트를 입력받아 데이터프레임을 생성
  if input_list[1]=='M':
    input_list.extend([0,1])
  else:
    input_list.extend([1,0])
  del input_list[1] 
  column_names = ['MESURE_AGE_CO','MESURE_IEM_001_VALUE','MESURE_IEM_002_VALUE','MESURE_IEM_003_VALUE','MESURE_IEM_005_VALUE','MESURE_IEM_006_VALUE','MESURE_IEM_012_VALUE','MESURE_IEM_018_VALUE','SEXDSTN_FLAG_CD_F','SEXDSTN_FLAG_CD_M']
  input_df = pd.DataFrame(np.array(input_list).reshape(1,10), index=["input"], columns=column_names)
  df = input_df.copy()
  prelist = []
  
  # 2) 분류 수행 함수 정의
  def fit_pred_pre (X_train, y_train, m_depth) :
    from sklearn.ensemble import RandomForestClassifier
    rf = RandomForestClassifier(max_depth=m_depth, n_estimators=400, random_state=42)
    rf.fit(X_train, y_train)
    pred = rf.predict(input_df)[0]
    colname='준비_'+y_train.name
    df[colname] = pred
    if pred == 1:
      prelist.append(colname)
  
  # 3) 분류 수행 함수 실행
  fit_pred_pre(X_train_over_pre1, y_train_over_pre1,6)
  fit_pred_pre(X_train_over_pre2, y_train_over_pre2,3)
  fit_pred_pre(X_train_over_pre3, y_train_over_pre3,3)
  fit_pred_pre(X_train_over_pre4, y_train_over_pre4,3)
  fit_pred_pre(X_train_over_pre5, y_train_over_pre5,7)
  fit_pred_pre(X_train_over_pre6, y_train_over_pre6,7)
  fit_pred_pre(X_train_over_pre7, y_train_over_pre7,4)
  fit_pred_pre(X_train_over_pre8, y_train_over_pre8,5)
  fit_pred_pre(X_train_over_pre9, y_train_over_pre9,3)
  fit_pred_pre(X_train_over_pre10, y_train_over_pre10,3)

  return(prelist, df)
```

- a, b, c 에서 준비/본/마무리 운동별 분류 함수의 경우 테스트용 피처 데이터셋을 개별 분류기와 동일하게 분류하는지 확인하기 위해 수치형 자료를 표준화시킨 값을 인자값으로 입력했으나, 최종 함수의 경우 원단위의 수치형 데이터를 리스트로 입력시 스케일링을 거쳐 분류를 진행하도록 코드를 작성했다.

```python
### 준비운동/본운동/마무리운동 분류 차례로 실행하는 최종함수

def total_df_list(user_input_list):
  # 1) 리스트를 입력받아 데이터프레임을 생성, 분류결과 저장 데이터프레임/리스트 생성
  user_input=pd.DataFrame({'MESURE_AGE_CO':[0],
                         'SEXDSTN_FLAG_CD':[0],
                         'MESURE_IEM_001_VALUE':[0],
                         'MESURE_IEM_002_VALUE':[0],
                         'MESURE_IEM_003_VALUE':[0],
                         'MESURE_IEM_005_VALUE':[0],
                         'MESURE_IEM_006_VALUE':[0],
                         'MESURE_IEM_012_VALUE':[0],
                         'MESURE_IEM_018_VALUE':[0]})

  # 리스트 형태로 입력받은 User data를 데이터 프레임 형태의 user_input에 저장
  for i in range(len(user_input_list)):
    user_input.iloc[0,i]=user_input_list[i]
  # 성별 제외 변수 standardiztion
  user_data_stand=standardization(user_input)
  # 성별 변수 더미화 
  user_data_stand=user_input_likeGetDummies(user_data_stand)

  input_df = user_data_stand
  df = input_df.copy()
  prelist = []
  mainlist = []
  endlist = []

  # 2) 분류 수행 함수 정의
  def fit_pred_pre (X_train, y_train, m_depth) :
    from sklearn.ensemble import RandomForestClassifier
    rf = RandomForestClassifier(max_depth=m_depth, n_estimators=400, random_state=42)
    rf.fit(X_train, y_train)
    pred = rf.predict(input_df)[0]
    colname='준비_'+y_train.name
    df[colname] = pred
    if pred == 1:
      prelist.append(colname)

  def fit_pred_main (X_train, y_train, m_depth) :
    from sklearn.ensemble import RandomForestClassifier
    rf = RandomForestClassifier(max_depth=m_depth, n_estimators=400, random_state=42)
    rf.fit(X_train, y_train)
    pred = rf.predict(input_df)[0]
    colname='본_'+y_train.name
    df[colname] = pred
    if pred == 1:
      mainlist.append(colname)

  def fit_pred_end (X_train, y_train, m_depth) :
    from sklearn.ensemble import RandomForestClassifier
    rf = RandomForestClassifier(max_depth=m_depth, n_estimators=400, random_state=42)
    rf.fit(X_train, y_train)
    pred = rf.predict(input_df)[0]
    colname='마무리_'+y_train.name
    df[colname] = pred
    if pred == 1:
      endlist.append(colname)
  
  # 전체 분류 수행 함수 실행
  fit_pred_pre(X_train_over_pre1, y_train_over_pre1,6)
  fit_pred_pre(X_train_over_pre2, y_train_over_pre2,3)
  fit_pred_pre(X_train_over_pre3, y_train_over_pre3,3)
  fit_pred_pre(X_train_over_pre4, y_train_over_pre4,3)
  fit_pred_pre(X_train_over_pre5, y_train_over_pre5,7)
  fit_pred_pre(X_train_over_pre6, y_train_over_pre6,7)
  fit_pred_pre(X_train_over_pre7, y_train_over_pre7,4)
  fit_pred_pre(X_train_over_pre8, y_train_over_pre8,5)
  fit_pred_pre(X_train_over_pre9, y_train_over_pre9,3)
  fit_pred_pre(X_train_over_pre10, y_train_over_pre10,3)

  fit_pred_main(X_train_over_main1, y_train_over_main1,7)
  fit_pred_main(X_train_over_main2, y_train_over_main2,3)
  fit_pred_main(X_train_over_main3, y_train_over_main3,5)
  fit_pred_main(X_train_over_main4, y_train_over_main4,3)
  fit_pred_main(X_train_over_main5, y_train_over_main5,7)
  fit_pred_main(X_train_over_main6, y_train_over_main6,4)
  fit_pred_main(X_train_over_main7, y_train_over_main7,3)
  fit_pred_main(X_train_over_main8, y_train_over_main8,4)
  fit_pred_main(X_train_over_main9, y_train_over_main9,6)
  fit_pred_main(X_train_over_main10, y_train_over_main10,3)
  fit_pred_main(X_train_over_main11, y_train_over_main11,7)
  fit_pred_main(X_train_over_main12, y_train_over_main12,3)
  fit_pred_main(X_train_over_main13, y_train_over_main13,7)
  fit_pred_main(X_train_over_main14, y_train_over_main14,3)
  fit_pred_main(X_train_over_main15, y_train_over_main15,10)
  fit_pred_main(X_train_over_main16, y_train_over_main16,3)
  fit_pred_main(X_train_over_main17, y_train_over_main17,3)
  fit_pred_main(X_train_over_main18, y_train_over_main18,6)
  fit_pred_main(X_train_over_main19, y_train_over_main19,4)
  fit_pred_main(X_train_over_main20, y_train_over_main20,7)
  fit_pred_main(X_train_over_main21, y_train_over_main21,9)
  fit_pred_main(X_train_over_main22, y_train_over_main22,3)
  fit_pred_main(X_train_over_main23, y_train_over_main23,3)
  fit_pred_main(X_train_over_main24, y_train_over_main24,7)
  fit_pred_main(X_train_over_main25, y_train_over_main25,3)
  fit_pred_main(X_train_over_main26, y_train_over_main26,3)
  fit_pred_main(X_train_over_main27, y_train_over_main27,3)
  fit_pred_main(X_train_over_main28, y_train_over_main28,10)
  fit_pred_main(X_train_over_main29, y_train_over_main29,10)
  fit_pred_main(X_train_over_main30, y_train_over_main30,3)
  fit_pred_main(X_train_over_main31, y_train_over_main31,9)
  fit_pred_main(X_train_over_main32, y_train_over_main32,3)
  fit_pred_main(X_train_over_main33, y_train_over_main33,4)
  fit_pred_main(X_train_over_main34, y_train_over_main34,3)
  fit_pred_main(X_train_over_main35, y_train_over_main35,6)
  fit_pred_main(X_train_over_main36, y_train_over_main36,7)

  fit_pred_end(X_train_over_end1, y_train_over_end1,3)
  fit_pred_end(X_train_over_end2, y_train_over_end2,3)
  fit_pred_end(X_train_over_end3, y_train_over_end3,4)
  fit_pred_end(X_train_over_end4, y_train_over_end4,5)
  fit_pred_end(X_train_over_end5, y_train_over_end5,4)
  fit_pred_end(X_train_over_end6, y_train_over_end6,3)
  fit_pred_end(X_train_over_end7, y_train_over_end7,7)
  fit_pred_end(X_train_over_end8, y_train_over_end8,3)

  # 4) 최종 반환값 생성
  total_list = prelist + mainlist + endlist
  total_df = df
              
  return (total_list, total_df)
```
